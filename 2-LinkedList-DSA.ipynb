```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
#Initial the head node 1st node creation
head = Node(10)

# Link to the second node
head.next = Node(20)

# Link to the 3rd Node
head.next.next = Node(30)

# Link to the 4th Node
head.next.next.next = Node(40)


# Print all the nodes
temp = head
while temp is not None:
    print(temp.data, end=" ")
    temp = temp.next

```
